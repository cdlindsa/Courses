```python
# Part 3 Neural Networks
# Please Note: The majority of code was adapted from SENG 474 Laboratory 2, Author Unknown, June 2020

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits, load_breast_cancer
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import numpy as np

# Preparing Dataset
# cleveland_hd = np.loadtxt('cleaned_processed.cleveland.data', delimiter = ',')
# size = cleveland_hd.shape[1]
# feature_names = ['age','sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
# class_names = ['no heart disease', 'heart disease']
# data = [i[0:size-1] for i in cleveland_hd]
# target = [i[size-1] for i in cleveland_hd]

# Preparing Dataset Breast Cancer Data
data, target = load_breast_cancer(return_X_y=True)
breast_cancer = load_breast_cancer()
feature_names = breast_cancer.feature_names
class_names = breast_cancer.target_names

X_scaled = preprocessing.scale(data)
trainData,testData,trainLabel,testLabel = train_test_split(X_scaled, target, test_size=0.2)

# 100 Iterations
clf_0 = MLPClassifier(hidden_layer_sizes=(1), max_iter=100, alpha=1e-4,solver='sgd', learning_rate_init=.01)
clf_0.fit(trainData,trainLabel)
print("Mean Accuracy: " + str(clf_0.score(trainData,trainLabel)))
print("Accuracy of using 1 neuron in the hidden layer: %0.4f",clf_0.score(testData,testLabel))
clf_1 = MLPClassifier(hidden_layer_sizes=(3), max_iter=100, alpha=1e-4,solver='sgd', learning_rate_init=.01)
clf_1.fit(trainData,trainLabel)
print("Mean Accuracy: " + str(clf_1.score(trainData,trainLabel)))
print("Accuracy of using 3 neurons in the hidden layer: %0.4f",clf_1.score(testData,testLabel))
clf_2 = MLPClassifier(hidden_layer_sizes=(10), max_iter=100, alpha=1e-4,solver='sgd', learning_rate_init=.01)
clf_2.fit(trainData,trainLabel)
print("Mean Accuracy: " + str(clf_2.score(trainData,trainLabel)))
print("Accuracy of using 10 neurons in the hidden layer: %0.4f",clf_2.score(testData,testLabel))
clf_3 = MLPClassifier(hidden_layer_sizes=(50), max_iter=100, alpha=1e-4,solver='sgd', learning_rate_init=.01)
clf_3.fit(trainData,trainLabel)
print("Mean Accuracy: " + str(clf_3.score(trainData,trainLabel)))
print("Accuracy of using 50 neurons in the hidden layer: %0.4f",clf_3.score(testData,testLabel))

# 500 Iterations
# clf_0 = MLPClassifier(hidden_layer_sizes=(1), max_iter=500, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_0.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_0.score(trainData,trainLabel)))
# print("Accuracy of using 1 neuron in the hidden layer: %0.4f",clf_0.score(testData,testLabel))
# clf_1 = MLPClassifier(hidden_layer_sizes=(3), max_iter=500, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_1.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_1.score(trainData,trainLabel)))
# print("Accuracy of using 3 neurons in the hidden layer: %0.4f",clf_1.score(testData,testLabel))
# clf_2 = MLPClassifier(hidden_layer_sizes=(10), max_iter=500, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_2.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_2.score(trainData,trainLabel)))
# print("Accuracy of using 10 neurons in the hidden layer: %0.4f",clf_2.score(testData,testLabel))
# clf_3 = MLPClassifier(hidden_layer_sizes=(50), max_iter=500, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_3.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_3.score(trainData,trainLabel)))
# print("Accuracy of using 50 neurons in the hidden layer: %0.4f",clf_3.score(testData,testLabel))

# 1000 Iterations
# clf_0 = MLPClassifier(hidden_layer_sizes=(1), max_iter=1000, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_0.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_0.score(trainData,trainLabel)))
# print("Accuracy of using 1 neuron in the hidden layer: %0.4f",clf_0.score(testData,testLabel))
# clf_1 = MLPClassifier(hidden_layer_sizes=(3), max_iter=1000, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_1.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_1.score(trainData,trainLabel)))
# print("Accuracy of using 3 neurons in the hidden layer: %0.4f",clf_1.score(testData,testLabel))
# clf_2 = MLPClassifier(hidden_layer_sizes=(10), max_iter=1000, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_2.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_2.score(trainData,trainLabel)))
# print("Accuracy of using 10 neurons in the hidden layer: %0.4f",clf_2.score(testData,testLabel))
# clf_3 = MLPClassifier(hidden_layer_sizes=(50), max_iter=1000, alpha=1e-4,solver='sgd', learning_rate_init=.01)
# clf_3.fit(trainData,trainLabel)
# print("Mean Accuracy: " + str(clf_3.score(trainData,trainLabel)))
# print("Accuracy of using 50 neurons in the hidden layer: %0.4f",clf_3.score(testData,testLabel))

plt.plot(clf_0.loss_curve_,'c--', label='1 Neurons')
plt.plot(clf_1.loss_curve_,'r--', label='3 Neurons')
plt.plot(clf_2.loss_curve_,'b--',label='10 Neurons')
plt.plot(clf_3.loss_curve_,'y--',label='50 Neurons')
plt.ylabel('loss', fontsize = 14)
plt.xlabel('n iterations', fontsize = 14)
plt.legend()
plt.show()
```
